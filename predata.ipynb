```python
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
from pandas_datareader import data as web
from datetime import datetime as dt
import quandl
```

```python
zips = 0
global zips
zips = pd.read_csv("ZIPSILVER.csv")

def house_price(code):
    str_code = str(code)
    df = quandl.get(
        'ZILLOW/Z'+str_code+'_MRP2B',
        authtoken="TgSaHF_YoDbd65fYqWkU"
        ).reset_index()
    #avg = (df.Value.iloc[:-150,]).mean
    avg_rent = round(df.Value.tail(3).mean())
    zips.loc[zips['ZipCode'] == code, 'avg_rent'] = avg_rent
```

```python
zips.head()
```

```python
import numpy as np
zipcodes = np.array(zips.ZipCode[(zips.StateCode == 'MO')
                        | (zips.StateCode == 'NY')
                        | (zips.StateCode == 'CA')
                        | (zips.StateCode == 'DC')
                        | (zips.StateCode == 'IN')
                        | (zips.StateCode == 'NJ')
                        | (zips.StateCode == 'WA')
                        | (zips.StateCode == 'MA')])
```

```python
len(zipcodes)
```

```python
for code in zipcodes:
    try:
        house_price(code)
        print('Successful')
    except:
        print('No zipcode')
```
